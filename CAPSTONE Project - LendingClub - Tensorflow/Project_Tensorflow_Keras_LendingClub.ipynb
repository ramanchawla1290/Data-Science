### The Data
___

We will be using a subset of the LendingClub DataSet obtained from Kaggle: https://www.kaggle.com/wordsforthewise/lending-club

LendingClub is a US peer-to-peer lending company, headquartered in San Francisco, California. It was the first peer-to-peer lender to register its offerings as securities with the Securities and Exchange Commission (SEC), and to offer loan trading on a secondary market. LendingClub is the world's largest peer-to-peer lending platform.

**Associated Files**

Data Features Info : **`lending_club_info.csv`**<br>
Actual Data Set    : **`lending_club_loans.csv`**


### Project Goal
___

Analysis of the historical data containing information on loans given out to borrowers and whether or not the borrower has defaulted (charge-off), we need to build a model that can predict whether a borrower will pay back their loan or not? This way in the future when the company gets a new potential customer, the model can assess whether or not the borrower is likely to pay back the loan.

The **`loan_status`** column contains our label.

**Filling missing values of `mort_acc` using the mean values of `total_acc`**

```python
def fill_mort_acc(total_acc, mort_acc):
    
    return round(total_acc_mean.loc[total_acc]) if np.isnan(mort_acc) else mort_acc
```

# Section 3 : Creating, Training and Evaluating our Model
___

### Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('fully_paid', axis=1).values

y = df['fully_paid'].values
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=101)
```

```python
X_train.shape
```

### Normalizing the Data

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# scaler = MinMaxScaler()

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

### Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(45, activation="relu"))
model.add(Dropout(0.25))

model.add(Dense(22, activation="relu"))
model.add(Dropout(0.25))

model.add(Dense(11, activation="relu"))
model.add(Dropout(0.25))

model.add(Dense(1, activation="sigmoid")) # Final Classification Node

model.compile(optimizer='adam', loss="binary_crossentropy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor="val_loss", mode="min", patience=25,
                           min_delta=0.01, restore_best_weights=True)
```

### Training the Model

```python
model.fit(X_train, y_train, epochs=100, batch_size=1000,
          validation_data=(X_test, y_test),
          callbacks=[early_stop], verbose=1)
```

### Saving the Model

```python
model.save('model_LendungClub.h5')
```

## Evaluation of Model

```python
losses = pd.DataFrame(model.history.history)

losses.plot(figsize=(9, 6))

plt.xlabel("Epochs")
```

### Model Predictions

```python
predictions = model.predict(X_test)

pred_classes = (predictions >= 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred_classes))
print()
print(classification_report(y_test, pred_classes))
```
